**Q.1. Write a python program which searches all the product under a particular product vertical from www.amazon.in. The product verticals to be searched will be taken as input from user. For e.g. If user input is ‘guitar’. Then search for guitars.**

```python
# connect to webdriver
driver = webdriver.Chrome('Downloads/chromedriver_win32/chromedriver.exe')
time.sleep(5)

# get www.amazon.in webpage
driver.get('https://www.amazon.in/')
time.sleep(5)

# get user input 
user_in = str(input('Enter product name :'))

# enter user input in search bar
product_search = driver.find_element_by_xpath('//input[@id="twotabsearchtextbox"]')
product_search.clear()
product_search.send_keys(user_in)
time.sleep(2)

# click the search button
search_butt = driver.find_element_by_xpath('//input[@id="nav-search-submit-button"]')
search_butt.click()
time.sleep(5)
```

**Q.2. In the above question, now scrape the following details of each product listed in first 3 pages of your search results and save it in a dataframe and csv. In case if any product vertical has less than 3 pages in search results then scrape all the products available under that product vertical. Details to be scraped are: "Brand Name", "Name of the Product", "Rating", "No. of Ratings", "Price", "Return/Exchange", "Expected Delivery", "Availability", "Other Details" and “Product URL”. In case, if any of the details are missing for any of the product then replace it by “-“.**

```python
# make empty lists to store details
br_name = []
pro_name = []
rate = []
total_rate = []
price = []
return_exch = []
expected_del = []
avail = []
other = []
pro_url = []

for page in range(3):  # scrape data from first 3 pages
    
    # get web elements of every product on current page
    data = driver.find_elements_by_xpath('//h2[@class="a-size-mini a-spacing-none a-color-base s-line-clamp-2"]/a[@class="a-link-normal a-text-normal" and @target="_blank"]')
    for d in data :
        end = 0
        try :  # try to click on product link
            d.click()
        except :
            try:   # if not able to click try using 'href' atrribute and then wait till the element is clickable
                ref = d.get_attribute("href")
                path = "//h2[@class="'a-size-mini a-spacing-none a-color-base s-line-clamp-2'"]/a[@href='"+ ref + "']"
                WebDriverWait(driver, 30).until(EC.element_to_be_clickable((By.XPATH,path))).click()
            except : 
                end = 1
        if end == 1: # if the product is not clickable at all, skip that product
            continue 
        time.sleep(5)
        driver.switch_to.window(driver.window_handles[-1]) # switch to new tab
        br_name.append(driver.find_element_by_xpath('//div[@class="a-section a-spacing-none"]/a[@id="bylineInfo" and @class="a-link-normal"]').text.strip('Brand: \n')) # append brand name
        pro_name.append(driver.find_element_by_xpath('//span[@id="productTitle" and @class="a-size-large product-title-word-break"]').text.strip('\n ')) # append product name
        try : # try to find ratings
            rate.append(driver.find_element_by_xpath('//span[@id="acrPopover" and @class="reviewCountTextLinkedHistogram noUnderline"]').get_attribute('title'))
        except : # if not found append '-'
            rate.append('-')
        try : # try to find total ratings
            total_rate.append(driver.find_element_by_xpath('//span[@id="acrCustomerReviewText" and @class="a-size-base"]').text.strip(' \n'))
        except : # else append '-'
            total_rate.append('-')
        try : # try to find price
            price.append(driver.find_element_by_xpath('//span[@id="priceblock_ourprice" or @id="priceblock_dealprice"]').text.strip(' \n'))
        except : # else append '-'
            price.append('-')
        try : # try to find replacement info
            return_exch.append(driver.find_element_by_xpath('//div[@id="RETURNS_POLICY"]//a[@class="a-size-small a-link-normal a-text-normal"]').text.strip(' \n'))
        except : # else append '-'
            return_exch.append('-')
        try : # try to find expected delivery date
            expected_del.append(driver.find_element_by_xpath('//div[@id="ddmDeliveryMessage" and @class="a-section a-spacing-mini"]//b').text.strip(' \n'))
        except : # else append '-'
            expected_del.append('-')
        try : # try to find product availability
            avail.append(driver.find_element_by_xpath('//div[@id="availability" and @class="a-section a-spacing-none"]/span[@class="a-size-medium a-color-success"]').text.strip(' \n'))
        except : # else append '-'
            avail.append('-')
        try : # try to find other details
            other.append(driver.find_element_by_xpath('//div[@id="feature-bullets" and @class="a-section a-spacing-medium a-spacing-top-small"]/ul[@class="a-unordered-list a-vertical a-spacing-mini"]').text.strip('\n '))
        except : # else append '-'
            other.append('-')
        pro_url.append(driver.current_url) # append product url
        driver.switch_to.window(driver.window_handles[0]) # switch to 1st tab
        time.sleep(5)
    try : # try to click on next page
        driver.find_element_by_xpath('//ul[@class="a-pagination"]//li[@class="a-last"]/a[text()="Next"]').click()
    except : # if not able to click, means there are less than 3 pages
        end = 2
    if end == 2 : # break the loop if less than 3 pages
        break
    time.sleep(5)

# dictionary
dict_ = {'Brand':br_name,'Product Name':pro_name,'Ratings':rate,'Total Ratings':total_rate,'Price':price,
        'Return/Exchange':return_exch,'Expected Delivery':expected_del,'Availabilty':avail,'Other Details':other,
         'Product URL':pro_url}
pd.set_option('display.max_rows',100)
df = pd.DataFrame(dict_)
df
```

```python
driver.quit()
```

**Q.3. Write a python program to access the search bar and search button on images.google.com and scrape 100 images each for keywords ‘fruits’, ‘cars’ and ‘Machine Learning’.**

```python
# connect to webdriver
driver = webdriver.Chrome('Downloads/chromedriver_win32/chromedriver.exe')
time.sleep(5)

# get images.google webpage
driver.get('https://images.google.com/')
time.sleep(5)

# create empty lists to store data
img = []
key_word = []

for word in ['fruits','cars','Machine Learning'] :
    
    # enter user input in search bar
    search = driver.find_element_by_xpath('//input[@class="gLFyf gsfi" or @class="JSAgYe"]')
    search.clear() # clear previous text present on search bar 
    search.send_keys(word) # send new text to search bar
    time.sleep(2)
    
    # click the search button
    try : 
        driver.find_element_by_xpath('//div[@class="FAuhyb"]/span[@class="z1asCe MZy1Rb"]').click()
    except :
        driver.find_element_by_xpath('//button[@class="KXJfe" and @id="BIqFsb"]/div[@class="OWOjU"]').click()
    time.sleep(5)
    
    # scroll down to a considerable length of page so that more than 100 web elements can be retrieved from the page
    driver.execute_script("window.scrollTo(0,5000)")   
    time.sleep(3)
    
    # get all the available web elements from the page
    data = driver.find_elements_by_xpath('//a[@class="wXeWr islib nfEiy mM5pbd"]/div[@class="bRMDJf islir"]')
    for d in data[0:100]: # scrape data from 1st 100 web elements
        driver.execute_script("arguments[0].scrollIntoView();",d) # scroll untill current element is visible for scrapping
        time.sleep(2)
        img.append(d.find_element_by_xpath('./img[@class="rg_i Q4LuWd"]').get_attribute('src')) # append image sorce code
        key_word.append(word) # append corresponding name of key word
    time.sleep(3)
# DataFrame
df = pd.DataFrame({'Key Word':key_word,'Image Download Link':img})
pd.set_option('display.max_rows',300)
df
```

```python
from pathlib import Path
Path("Desktop/Assignment_Images").mkdir()  # create an empty folder to store images
```

```python
import urllib
for i,key in enumerate(df['Key Word']) :  # download all images one by one and store them into above created folder
    x = key + '_' + str(i+1) + '.jpeg'
    urllib.request.urlretrieve(df['Image Download Link'][i], "Desktop/Assignment_Images/"+ x)
    time.sleep(1)
```

**Q.4. Write a python program to search for a smartphone(e.g.: Oneplus Nord, pixel 4A, etc.) on www.flipkart.com and scrape following details for all the search results displayed on 1st page. Details to be scraped: “Brand Name”, “Smartphone name”, “Colour”, “RAM”, “Storage(ROM)”, “Primary Camera”, “Secondary Camera”, “Display Size”, “Display Resolution”, “Processor”, “Processor Cores”, “Battery Capacity”, “Price”, “Product URL”.
```python
# connect to webdriver
driver = webdriver.Chrome('Downloads/chromedriver_win32/chromedriver.exe')
time.sleep(5)

# get flipkart webpage
driver.get('https://www.flipkart.com/')
time.sleep(5)

# close login box if asked
try :
    driver.find_element_by_xpath('//div[@class="_2QfC02"]//button[@class="_2KpZ6l _2doB4z"]').click()
except :
    pass

# get user input 
user_in = str(input('Enter smartphone name :'))

# enter user input in search bar
search = driver.find_element_by_xpath('//div[@class="_3OO5Xc"]/input[@class="_3704LK"]')
search.clear()
search.send_keys(user_in)
time.sleep(2)

# click the search button
search_butt = driver.find_element_by_xpath('//div[@class="col-12-12 _2oO9oE"]//button[@class="L0Z3Pu"]')
search_butt.click()
time.sleep(5)

# get web elements of every product on 1st page
data = driver.find_elements_by_xpath('//div[@class="col col-7-12"]/div[@class="_4rR01T"]')

# make empty lists to store details
br_name = []
ph_name = []
clr = []
ram = []
rom = []
pri_camera = []
sec_camera = []
display_size = []
display_reso = []
processor = []
processor_cores = []
battery = []
price = []
pro_url = []

# scrape data 
for d in data :
    d.click()
    time.sleep(5)
    driver.switch_to.window(driver.window_handles[-1]) # switch to new tab
    driver.find_element_by_xpath('//button[@class="_2KpZ6l _1FH0tX"]').click() # click on ream more button of specifications
    time.sleep(2)
    name = driver.find_element_by_xpath('//h1[@class="yhB1nd"]/span[@class="B_NuCI"]').text.strip('\n ')
    br_name.append(name.split()[0]) # append brand name
    ph_name.append(driver.find_element_by_xpath('//div[@class="_3k-BhJ"]//td[contains(text(),"Model Name")]/following-sibling::td[@class="URwL2w col col-9-12"]').text.strip('\n ')) #append model name
    try : # try to find color
        clr.append(driver.find_element_by_xpath('//div[@class="_3k-BhJ"]//td[contains(text(),"Color")]/following-sibling::td[@class="URwL2w col col-9-12"]').text.strip('\n ')) #append color 
    except : # if no color found, append '-'
        clr.append('-')
    try : # try to find display size
        display_size.append(driver.find_element_by_xpath('//div[@class="_3k-BhJ"]//td[contains(text(),"Display Size")]/following-sibling::td[@class="URwL2w col col-9-12"]').text.strip('\n '))  #append display size
    except : # if display size not found, append '-'
        display_size.append('-')
    try :  # try to find resolution
        display_reso.append(driver.find_element_by_xpath('//div[@class="_3k-BhJ"]//td[contains(text(),"Resolution")]/following-sibling::td[@class="URwL2w col col-9-12"]').text.strip('\n ')) #append resolution
    except : # if no resolution found, append '-'
        display_reso.append('-')
    try : # try finding processor
        processor.append(driver.find_element_by_xpath('//div[@class="_3k-BhJ"]//td[contains(text(),"Processor Type")]/following-sibling::td[@class="URwL2w col col-9-12"]').text.strip('\n ')) # append processor
    except : # if no processor found
        processor.append('-')
    try : # try to find cores
        processor_cores.append(driver.find_element_by_xpath('//div[@class="_3k-BhJ"]//td[contains(text(),"Processor Core")]/following-sibling::td[@class="URwL2w col col-9-12"]').text.strip('\n ')) # append cores
    except : # if no cores found, append '-'
        processor_cores.append('-')
    try : # try finding storage(rom)
        rom.append(driver.find_element_by_xpath('//div[@class="_3k-BhJ"]//td[contains(text(),"Internal Storage")]/following-sibling::td[@class="URwL2w col col-9-12"]').text.strip('\n ')) # append storage(rom)
    except : # if not found append '-'
        rom.append('-')
    try : # try finding ram
        ram.append(driver.find_element_by_xpath('//div[@class="_3k-BhJ"]//td[contains(text(),"RAM")]/following-sibling::td[@class="URwL2w col col-9-12"]').text.strip('\n ')) # append ram
    except : # if not found append '-'
        ram.append('-')
    try : # try finding primary camera
        pri_camera.append(driver.find_element_by_xpath('//div[@class="_3k-BhJ"]//td[text()="Primary Camera"]/following-sibling::td[@class="URwL2w col col-9-12"]').text.strip('\n ')) # append primary cam
    except : # if not found append '-'
        pri_camera.append('-')
    try : # try finding secondary camera 
        sec_camera.append(driver.find_element_by_xpath('//div[@class="_3k-BhJ"]//td[text()="Secondary Camera"]/following-sibling::td[@class="URwL2w col col-9-12"]').text.strip('\n ')) # append secondary cam
    except : # if not found append '-'
        sec_camera.append('-')
    pro_url.append(driver.current_url)  # append product url
    price.append(driver.find_element_by_xpath('//div[@class="_25b18c"]/div[@class="_30jeq3 _16Jk6d"]').text.strip('\n ')) # append price
    try : # try finding battery capacity
        battery.append(driver.find_element_by_xpath('//div[@class="_3k-BhJ"]//td[contains(text(),"Battery Capacity")]/following-sibling::td[@class="URwL2w col col-9-12"]').text.strip('\n ')) # append battery capacity
    except : # if not found append '-'
        battery.append('-')
    driver.switch_to.window(driver.window_handles[0]) # switch to 1st tab
    time.sleep(5)

# dictionary
dict_ = {'Brand':br_name,'Name of Model':ph_name,'Color':clr,'RAM':ram,'Storage(ROM)':rom,'Primary Camera':pri_camera,
        'Secondary Camera':sec_camera,'Display Size':display_size,'Display Resolution':display_reso,'Processor':processor,
        'Processor Cores':processor_cores,'Battery Capacity':battery,'Price':price,'Product URL':pro_url}
df = pd.DataFrame(dict_)
df
```

**Q.6. Write a program to scrap details of all the funding deals for second quarter (i.e. July 20 – September 20) from trak.in.**

```python
# connect to webdriver
driver = webdriver.Chrome('Downloads/chromedriver_win32/chromedriver.exe')
driver.maximize_window()
time.sleep(5)

# get trak.in webpage
driver.get('https://trak.in')
time.sleep(5)

# click on fund deals option
driver.find_element_by_xpath('//a[@title="https://trak.in/india-startup-funding-investment-2015/"]').click()
time.sleep(20)

# show maximum entries of last 1st 12 tables
for i in driver.find_elements_by_xpath('//*[text()="100"]')[0:12] :
    i.click()
    time.sleep(2)

# empty lists to store data
date = []
name = []
ind_vert = []
sub_vert = []
city = []
invester = []
invest_type = []
amount = []

# scrape data
for clk in driver.find_elements_by_xpath('//h2') :
    if clk.text in ["July, 2020","August, 2020","September, 2020"] :
    
        data = clk.find_element_by_xpath('./following-sibling::div[@class="dataTables_wrapper no-footer"]//tbody[@class="row-hover"]')
        
        for d in data.find_elements_by_xpath('.//tr[@role="row"]') :
            
            date.append(d.find_element_by_xpath('.//td[2]').text.strip('\n ')) #append date
            
            name.append(d.find_element_by_xpath('.//td[3]').text.strip('\n '))  # append name of start up
            
            ind_vert.append(d.find_element_by_xpath('.//td[4]').text.strip('\n ')) # append industry type
            
            sub_vert.append(d.find_element_by_xpath('.//td[5]').text.strip('\n ')) # append industry subtype
            
            city.append(d.find_element_by_xpath('.//td[6]').text.strip('\n ')) # append city name
            
            invester.append(d.find_element_by_xpath('.//td[7]').text.strip('\n ')) # append name of invester
            
            invest_type.append(d.find_element_by_xpath('.//td[8]').text.strip('\n ')) # append type of investment
            
            amount.append(d.find_element_by_xpath('.//td[9]').text.strip('\n ')) # append amount of investment

# Data frame
dict_ = {'Date':date,'Start Up Name':name,'Industry/Vertical':ind_vert,'Sub-Vertical':sub_vert,'City':city,
       'Invester Name':invester,'Investment type':invest_type,'Amount(in USD)':amount}
df = pd.DataFrame(dict_)

df
```

```python
driver.quit()
```

**Q.7. Write a program to scrap all the available details of top 10 gaming laptops from digit.in.**

```python
# connect to webdriver
driver = webdriver.Chrome('Downloads/chromedriver_win32/chromedriver.exe')
driver.maximize_window()
time.sleep(5)

# get https://www.digit.in/ webpage
driver.get('https://www.digit.in/')
time.sleep(5)

# click on top 10 best gaming laptops link
driver.find_element_by_xpath('//a[@href="https://www.digit.in/top-products/best-gaming-laptops-40.html"]').click()
time.sleep(5)

# empty lists to store data
name = []
seller = []
price = []
os = []
processor = []
display = []
memory = []

# scrape data
data = driver.find_elements_by_xpath('//table[@id="summtable"]/tbody//tr')
for d in data[1:] :
    name.append(d.find_element_by_xpath('.//td[1]').text.strip('\n '))  # append name of product
    seller.append(d.find_element_by_xpath('.//td[2]').text.strip('\n '))  # apend name of seller
    price.append(d.find_element_by_xpath('.//td[3]').text.strip('\n ')) # append price

extra = driver.find_elements_by_xpath('//div[@class="Specs-Wrap"]/ul')
for e in extra :
    os.append(e.find_element_by_xpath('.//li[1]//div[@class="Wrap-pop"]').text.strip('OS\n ')) #append OS
    display.append(e.find_element_by_xpath('.//li[2]//div[@class="Wrap-pop"]').text.strip('DISPLAY\n ')) #append display details
    processor.append(e.find_element_by_xpath('.//li[3]//div[@class="Wrap-pop"]').text.strip('PROCESSOR\n ')) # append processor 
    memory.append(e.find_element_by_xpath('.//li[4]//div[@class="Wrap-pop"]').text.strip('MEMORY\n ')) # append memory details

# create data frame
df = pd.DataFrame({'Product Name':name,'Seller':seller,'Price(Rs.)':price,
                  'OS':os,'Display':display,'Processor':processor,'Memory':memory})
df
```
