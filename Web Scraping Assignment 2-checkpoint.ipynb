```python
# connect to webdriver
driver = webdriver.Chrome('Downloads/chromedriver_win32/chromedriver.exe')
```

```python
# get Naukri.com webpage
driver.get('https://www.naukri.com/')
time.sleep(10)
```

```python
# enter 'Data Scientist' in job search bar 
job_search = driver.find_element_by_xpath('//input[@id="qsb-keyword-sugg"]')
job_search.send_keys('Data Scientist')

# enter 'Banglore' in location search bar
loc_search = driver.find_element_by_xpath('//input[@id="qsb-location-sugg"]')
loc_search.send_keys('Banglore')
```

```python
# click the search button
search_butt = driver.find_element_by_xpath('//div[@class="search-btn"]/button')
search_butt.click()
time.sleep(10)
```

```python
# scrape the data for first 10 jobs results

# scrape job_title
jobs = driver.find_elements_by_xpath('//a[@class="title fw500 ellipsis"]') 
job_title = [job.text for job in jobs]

# scrape job_location
locs = driver.find_elements_by_xpath('//li[@class="fleft grey-text br2 placeHolderLi location"]') 
job_locs = [loc.text for loc in locs]

# scrape company name
co_name = driver.find_elements_by_xpath('//div[@class="info fleft"]//div[@class="mt-7 companyInfo subheading lh16"]')
names = [name.text for name in co_name]

# scrape job description
desc = driver.find_elements_by_xpath('//div[@class="info fleft"]/a')
jds = []
for url in desc[0:10]:
    error = 0
    try :
        url.click()
    except :
        try:
            ref = url.get_attribute("href")
            path = "//div[@class="'info fleft'"]//a[@href='"+ ref + "']"
            WebDriverWait(driver, 30).until(EC.element_to_be_clickable((By.XPATH,path))).click()
        except:
            error = 1
    if error == 1:
        jds.append('ElementClickInterceptedException & then TimeoutException, could not extract job description')
    else :
        time.sleep(10)
        driver.switch_to.window(driver.window_handles[-1]) # switch to new tab
        try :
            jd = driver.find_element_by_xpath('//section[@class="job-desc"]')
        except :
            jds.append('Job Description webpage not from naukri.com, could not extract job description')
            error = 2
        if error !=2 :
            jds.append(jd.text)
        driver.switch_to.window(driver.window_handles[0]) # switch to 1st tab
    time.sleep(5)
```

```python
# save scraped data in a dataFrame
df_jobs_desc = pd.DataFrame({'Job Title':job_title[0:10],'Job Location':job_locs[0:10],'Company Name':names[0:10],
                        'Job Description':jds})
pd.set_option('display.max_colwidth', None)
df_jobs_desc
```

```python
# connect to webdriver
driver = webdriver.Chrome('Downloads/chromedriver_win32/chromedriver.exe')
```

```python
# get webpage
driver.get('https://www.glassdoor.co.in/Job/index.htm')
```

```python
# enter 'Data Scientist' in job search bar 
job_search = driver.find_element_by_xpath('//input[@id="KeywordSearch"]')
job_search.send_keys('Data Scientist')

# enter 'Noida' in location search bar
loc_search = driver.find_element_by_xpath('//input[@id="LocationSearch"]')
loc_search.clear()
loc_search.send_keys('Noida')
```

```python
# click the search button
search_butt = driver.find_element_by_xpath('//button[@class="gd-btn-mkt"]')
search_butt.click()
time.sleep(10)
```

```python
# scrape data
whole = driver.find_elements_by_xpath('//li[@class="react-job-listing css-wp148e eigr9kq3"]')

names = []
days = []
ratings = []

for w in whole[0:10] :
    
    error = 0
    
    # scrape company name
    co_name = w.find_element_by_xpath('.//div[@class="d-flex justify-content-between align-items-start"]')
    names.append(co_name.text)

    # scrape total days after job posting
    job_age = w.find_element_by_xpath('.//div[@data-test="job-age"]')
    days.append(job_age.text.strip('d'))

    # scrape company rating
    try :
        stars = w.find_element_by_xpath('.//div[@class="d-flex flex-column css-x75kgh e1rrn5ka3"]//span[@class="css-19pjha7 e1cjmv6j1"]')
    except :
        error = 1
    if error == 1 :             # if no rating is present append None to the list
        ratings.append(None)
    else :
        ratings.append(stars.text)
```

```python
# save scraped data in a dataFrame
df_jobs_gd = pd.DataFrame({'Company Name':names[0:10],'Posted number of days ago':days[0:10],
                        'Company Ratings':ratings[0:10]})
df_jobs_gd
```

**Q6 : Scrape data of first 100 sunglasses listings on flipkart.com. You have to scrape four attributes:**

**1. Brand**
**2. Product Description**
**3. Price**
**4. Discount %**

```python
# connect to webdriver
driver = webdriver.Chrome('Downloads/chromedriver_win32/chromedriver.exe')
```

```python
# get webpage
driver.get('https://www.flipkart.com/')
time.sleep(10)
```

```python
# click to close sign in pop-up if it asks to sign in else pass
try :
    driver.find_element_by_xpath('//div[@class="_2QfC02"]/button[@class="_2KpZ6l _2doB4z"]').click()
except :
    pass
```

```python
# enter 'Sunglasses' in search bar 
search = driver.find_element_by_xpath('//input[@class="_3704LK"]')
search.send_keys('sunglasses')

time.sleep(3)

# click the search button
search_butt = driver.find_element_by_xpath('//button[@class="L0Z3Pu"]')
search_butt.click()
time.sleep(10)
```

```python
# scrape data

brands = []
product_desc = []
prices = []
discount = []

while len(brands) < 100 :
    
    whole = driver.find_elements_by_xpath('//div[@class="_1xHGtK _373qXS"]//div[@class="_2B099V"]')
    for w in whole :
        
        while len(prices) < 100 :

            error = 0

            # scrape brand name
            br_name = w.find_element_by_xpath('.//div[@class="_2WkVRV"]')
            brands.append(br_name.text)

            # scrape product description
            product_d = w.find_element_by_xpath('.//a[@class="IRpwTa" or @class="IRpwTa _2-ICcC"]')
            product_desc.append(product_d.text)

            # scrape prices
            prcs = w.find_element_by_xpath('.//div[@class="_30jeq3"]')
            prices.append(prcs.text)

            # scrape discounts
            try :
                dis = w.find_element_by_xpath('.//div[@class="_3Ay6Sb"]')
            except :
                error = 1 
            if error == 1 :            # if no discount is present, append None to the list
                discount.append(None)
            else :
                discount.append(dis.text)
            
            break
    # click on next page
    driver.find_element_by_xpath('//div[@class="_2MImiq"]//a[@class="_1LKTO3"]/span[contains(text(),"Next")]').click()
    time.sleep(10)
```

```python
# save scraped data in a dataFrame
pd.set_option('display.max_rows',100)
df_sunglasses = pd.DataFrame({'Brands':brands,'Product Description':product_desc,
                        'Prices':prices,'Discounts':discount})
df_sunglasses
```

```python
driver.quit()
```

You have to go the link: https://www.flipkart.com/apple-iphone-11-black-64-gb-includes-earpods-power-adapter/p/itm0f37c2240b217?pid=MOBFKCTSVZAXUHGR&lid=LSTMOBFKCTSVZAXUHGREPBFGI&marketplace.**


```python
# connect to webdriver
driver = webdriver.Chrome('Downloads/chromedriver_win32/chromedriver.exe')
```

```python
# get webpage
driver.get('https://www.flipkart.com/apple-iphone-11-black-64-gb-includes-earpods-power-adapter/p/itm0f37c2240b217?pid=MOBFKCTSVZAXUHGR&lid=LSTMOBFKCTSVZAXUHGREPBFGI&marketplace')
```

```python
# click on 'ALL REVIEWS' button.
driver.find_element_by_xpath('//div[@class="_3UAT2v _16PBlm"]/span').click()
time.sleep(5)
```

```python
# scrape data of first 100 reviews

rev_sum = []
full_rev = []
ratings = []

while len(rev_sum) < 100 :
    
    # click on 'read more' buttons.
    read_mores = driver.find_elements_by_xpath('//div[@class="col _2wzgFH K0kLPL"]//span[@class="_1BWGvX"]')
    for read in read_mores :
        try : # click only if read more option is present otherwise pass
            read.find_element_by_xpath('./span[contains(text(),"READ MORE")]').click()
        except :
            pass
        time.sleep(5)
    
    whole = driver.find_elements_by_xpath('//div[@class="col _2wzgFH K0kLPL"]')
    for w in whole :
        
        while len(full_rev) < 100 :

            # scrape ratings
            stars = w.find_element_by_xpath('.//div[@class="_3LWZlK _1BLPMq" or @class="_3LWZlK _1rdVr6 _1BLPMq"]')
            ratings.append(stars.text.strip('\n'))

            # scrape review summary
            summ = w.find_element_by_xpath('.//p[@class="_2-N8zT"]')
            rev_sum.append(summ.text)

            # scrape full review
            full = w.find_element_by_xpath('.//div[@class="t-ZTKy"]/div/div')
            full_rev.append(full.text)
            
            break
    # click on next page    
    driver.find_element_by_xpath('//nav[@class="yFHi8N"]//a[@class="_1LKTO3"]/span[contains(text(),"Next")]').click()
    time.sleep(10)
```

```python
# save scraped data in a dataFrame
pd.set_option('display.max_rows',100)
df_reviews = pd.DataFrame({'Ratings':ratings,'Review Summary':rev_sum,'Full Review':full_rev})
df_reviews
```

```python
# connect to webdriver
driver = webdriver.Chrome('Downloads/chromedriver_win32/chromedriver.exe')
```

```python
# get myntra shoes webpage
driver.get('https://www.myntra.com/shoes')
time.sleep(10)
```

```python
# check the filter option near to 'Rs. 6649 to Rs. 13099' filter option,
# as the price filter options keeps on changing dynamically
fltr_opt = (driver.find_element_by_xpath('//ul[@class="price-list"]//li[2]//input')).get_attribute('value')
print('filter option of Rs.',fltr_opt,'is available instead of filter option of Rs. 6649 to 13099')

# select check box of price filter 
chk_prc = driver.find_element_by_xpath('//ul[@class="price-list"]//li[2]//input[@value="'+fltr_opt+'"]/following-sibling::div[@class="common-checkboxIndicator"]')
chk_prc.click()

time.sleep(3)

# select check box of 'black' in color filter
chk_clr = driver.find_element_by_xpath('//li[@class="colour-listItem"]//input[@value="Black"]/following-sibling::div[@class="common-checkboxIndicator"]')
chk_clr.click()
time.sleep(10)
```

```python
# scrape data of first 100 shoes

brand = []
short_desc = []
price = []

while len(brand) < 100 :
    
    whole = driver.find_elements_by_xpath('//li[@class="product-base"]//div[@class="product-productMetaInfo"]')
    for w in whole :
        
        while len(price) < 100 :

            # scrape brands
            brnd = w.find_element_by_xpath('.//h3[@class="product-brand"]')
            brand.append(brnd.text)

            # scrape short description
            shrt = w.find_element_by_xpath('.//h4[@class="product-product"]')
            short_desc.append(shrt.text)

            # scrape prices 
            try :  #  scrape with following element if product has a discount
                prcs = w.find_element_by_xpath('.//div[@class="product-price"]//span[@class="product-discountedPrice"]')
            except : # scrape with following element if product does not have a discount
                prcs = w.find_element_by_xpath('.//div[@class="product-price"]/span')
            price.append(prcs.text.strip(' \nRs.'))
            
            break
    # click on next page  
    driver.find_element_by_xpath('//ul[@class="pagination-container"]//li[@class="pagination-next"]/a[contains(text(),"Next")]').click()
    time.sleep(10)
```

```python
# connect to webdriver
driver = webdriver.Chrome('Downloads/chromedriver_win32/chromedriver.exe')
```

```python
# get amazon webpage
driver.get('https://www.amazon.in')
time.sleep(10)

# enter 'laptop' in search bar 
search = driver.find_element_by_xpath('//input[@id="twotabsearchtextbox"]')
search.send_keys('Laptop')

time.sleep(3)

# click the search button
search_butt = driver.find_element_by_xpath('//input[@id="nav-search-submit-button"]')
search_butt.click()
time.sleep(10)
```

```python
# select check box of 'intel core i7' in CPU filter 
chk_i7 = driver.find_element_by_xpath('//div[@id="filters"]//li[@aria-label="Intel Core i7"]//a[@class="a-link-normal s-navigation-item"]')
chk_i7.click()

time.sleep(5)

# select check box of 'intel core i9' in CPU filter
chk_i9 = driver.find_element_by_xpath('//div[@id="filters"]//li[@aria-label="Intel Core i9"]//a[@class="a-link-normal s-navigation-item"]')
chk_i9.click()

time.sleep(5)
```

```python
# scrape data of first 10 laptops on the page

ttl = []
ratings = []
price = []

# scrape titles
title = driver.find_elements_by_xpath('//div[@class="sg-col-inner"]//span[@class="a-size-medium a-color-base a-text-normal"]')
for t in title[0:10] :
    ttl.append(t.text)

# scrape ratings
stars = driver.find_elements_by_xpath('//div[@class="sg-col-inner"]//div[@class="a-row a-size-small"]/span')
for s in stars[0:20:2] :
    ratings.append(s.get_attribute('aria-label'))

# scrape prices
prcs = driver.find_elements_by_xpath('//div[@class="sg-col-inner"]//span[@aria-hidden="true"]//span[@class="a-price-whole"]')
for p in prcs[0:10]:
    price.append(p.text)
```

```python
# save scraped data in a dataFrame
pd.set_option('display.max_rows',10)
pd.set_option('display.max_colwidth',None)
df_laptops = pd.DataFrame({'Titles':ttl,'Ratings':ratings,
                        'Prices(Rs.)':price})
df_laptops
```
