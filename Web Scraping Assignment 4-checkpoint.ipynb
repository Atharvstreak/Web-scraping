**Q.1. Scrape the details of most viewed videos on YouTube from Wikipedia:
Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos/**

**You need to find following details:**

**A) Rank**

**B) Name**

**C) Artist**

**D) Upload date**

**E) Views**

```python
# connect to webdriver
driver = webdriver.Chrome('Downloads/chromedriver_win32/chromedriver.exe')
time.sleep(5)

# get webpage
driver.get('https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos')
time.sleep(5)

# create empty lists to store data
rank = []
name = []
artist = []
up_date = []
views = []

# scrape data
data = driver.find_elements_by_xpath('//div[@id="mw-content-text"]//table[@class="wikitable sortable jquery-tablesorter"][1]//tbody//tr')
for i,d in enumerate(data,start=1) :
    rank.append(i)
    n = d.find_element_by_xpath('.//td[2]').text.strip('\n "')
    name.append(re.findall('(.+)"',n)[0])
    artist.append(d.find_element_by_xpath('.//td[3]').text.strip('\n '))
    up_date.append(d.find_element_by_xpath('.//td[5]').text.strip('\n '))
    views.append(d.find_element_by_xpath('.//td[4]').text.strip('\n '))
    
# data recorded on the date
print('Data',driver.find_element_by_xpath('//div[@id="mw-content-text"]//table[@class="wikitable sortable jquery-tablesorter"][1]//tfoot').text.strip('\n '))

# dataframe
df = pd.DataFrame({'Rank':rank,'Name':name,'Artist/Uploader':artist,'Upload Date':up_date,'Total Views(Billions)':views})
df
```

```python
driver.quit()
```

**Q.2. Scrape the details team India’s international fixtures from bcci.tv.
Url = https://www.bcci.tv/.**

**You need to find following details:**

**A) Match title (I.e. 1st ODI)**

**B) Series**

**C) Place**

**D) Date**

**E) Time**

**Note: - From bcci.tv home page you have reach to the international fixture page through code.**

```python
# connect to webdriver
driver = webdriver.Chrome('Downloads/chromedriver_win32/chromedriver.exe')
time.sleep(5)

# get webpage
driver.get('https://www.bcci.tv/')
time.sleep(5)

# navigate to international section
clk = driver.find_element_by_xpath('//ul[@class="navigation__list showMoreEnabled"]//div[@class="navigation__drop-down drop-down drop-down--reveal-on-hover"]')
clk.click()
time.sleep(1)

# click on fixtures
clk.find_element_by_xpath('.//li[@class="drop-down__option  "][1]').click()
time.sleep(3)

# create empty lists to store data
ttl = []
series = []
place = []
date = []
time_ = []

# scrape data
data = driver.find_elements_by_xpath('//div[@class="js-list"]//a')
for d in data :
    ttl.append(d.find_element_by_xpath('.//p[@class="fixture__additional-info"]/strong').text.strip('\n '))
    place.append(d.find_element_by_xpath('.//p[@class="fixture__additional-info"]//span').text.strip('\n '))
    temp = d.find_element_by_xpath('.//div[@class="fixture__datetime desktop-only"]').text.strip('\n ').split('\n')
    date.append((temp[0]+' '+temp[1]+' '+temp[2]).strip('\n '))
    time_.append(temp[-1].strip('\n '))
    series.append(d.find_element_by_xpath('.//div[@class="fixture__format-strip"]//span[2]').text.strip('\n '))
    
# dataframe
df = pd.DataFrame({'Match title':ttl,'Series':series,'Place':place,'Date':date,'Time':time_})
df
```

**Q.7. Scrape the details of Data science recruiters from naukri.com.
Url = https://www.naukri.com/**

**You have to find the following details:**

**A) Name**

**B) Designation**

**C) Company**

**D) Skills they hire for**

**E) Location**

**Note: - From naukri.com homepage click on the recruiters option and the on the search pane type Data science and click on search. All this should be done through code**

```python
# connect to webdriver
driver = webdriver.Chrome('Downloads/chromedriver_win32/chromedriver.exe')
time.sleep(5)

# get webpage
driver.get('https://www.naukri.com/')
time.sleep(5)

# close all ad windows
for w in driver.window_handles[1:] :
    driver.switch_to.window(w)
    driver.close()
    time.sleep(1)

# click on recruiters
driver.switch_to.window(driver.window_handles[0])
driver.find_element_by_xpath('//ul[@class="midSec menu"]//div[contains(text(),"Recruiters")]').click()
time.sleep(3)

# input data science in search bar
driver.switch_to.window(driver.window_handles[-1])
driver.find_element_by_xpath('//div[@class="inpWrap"]/input[@class="sugInp"]').send_keys('Data Science')
time.sleep(1)

# click on search button
driver.find_element_by_xpath('//button[@class="fl qsbSrch blueBtn"]').click()
time.sleep(3)

# create empty lists to store data 
name = []
des = []
co = []
loc = []
skills = []

# scrape data 
last = 0
while last != 1 : # when last = 1, break loop 
    data_odd = driver.find_elements_by_xpath('//div[@class="recSec fl fadeInUp"]') # collect half data on page 
    data_even = driver.find_elements_by_xpath('//div[@class="recSec fr fadeInUp"]')  # collect other half data on page
    o = 0 # initialise index variable for data_odd
    e = 0 # initialise index variable for data_even
    for i in range(len(data_odd) + len(data_even)):
        if i%2 == 0: # if the loop number is even append data from data_odd
            try :
                name.append(data_odd[o].find_element_by_xpath('.//p[@class="highlightable"]//a[1]').text.strip('\n '))
            except :
                name.append('-')
            try :
                co.append(data_odd[o].find_element_by_xpath('.//p[@class="highlightable"]//a[2]').text.strip('\n '))
            except :
                co.append('-')
            try :
                des.append(data_odd[o].find_element_by_xpath('.//p[@class="highlightable"]//span[1]').text.strip('\n '))
            except :
                co.append('-')
            try :
                loc.append(data_odd[o].find_element_by_xpath('.//p[@class="highlightable"]//span[2]').text.strip('\n '))
            except :
                loc.append('-')
            try :
                skills.append(data_odd[o].find_element_by_xpath('.//div[@class="hireSec highlightable"]').text.strip('\n '))
            except :
                skills.append('-')
            o += 1
        else : # else append data from data_even
            try :
                name.append(data_even[e].find_element_by_xpath('.//p[@class="highlightable"]//a[1]').text.strip('\n '))
            except :
                name.append('-')
            try :
                co.append(data_even[e].find_element_by_xpath('.//p[@class="highlightable"]//a[2]').text.strip('\n '))
            except :
                co.append('-')
            try :
                des.append(data_even[e].find_element_by_xpath('.//p[@class="highlightable"]//span[1]').text.strip('\n '))
            except :
                co.append('-')
            try :
                loc.append(data_even[e].find_element_by_xpath('.//p[@class="highlightable"]//span[2]').text.strip('\n '))
            except :
                loc.append('-')
            try :
                skills.append(data_even[e].find_element_by_xpath('.//div[@class="hireSec highlightable"]').text.strip('\n '))
            except :
                skills.append('-')
            e += 1
    try :
        driver.find_element_by_xpath('//button[@class="grayBtn" and contains(text(),"Next")]').click()
    except : # assign 1 to 'last' when reached at last page
        last = 1
    time.sleep(5)

# dataframe
df = pd.DataFrame({'Recruiter Name':name,'Designation':des,'Company':co,'Location':loc,'Skills they hire for':skills})
pd.set_option('display.max_colwidth',None)
df
```

```python
driver.quit()
```

**Q.10. Details of Datasets from UCI machine learning repositories.
Url = https://archive.ics.uci.edu/**

**You have to find the following details:**

**A) Dataset name**

**B) Data type**

**C) Task**

**D) Attribute type**

**E) No of instances**

**F) No of attribute**

**G) Year**

**Note: - from the home page you have to go to the ShowAllDataset page through code.**

```python
# connect to webdriver
driver = webdriver.Chrome('Downloads/chromedriver_win32/chromedriver.exe')
time.sleep(5)

# get webpage
driver.get('https://archive.ics.uci.edu/ml/index.php')
time.sleep(5)

# click on view all data sets
driver.find_element_by_xpath('//span[@class="whitetext"]//b[contains(text(),"View ALL Data Sets")]').click()
time.sleep(5)

# create empty lists to store data
name = []
dtype = []
task = []
atype = []
inst = []
attri = []
year = []

# scrape data
data = driver.find_elements_by_xpath('//td[@valign="top"][2]//tbody//tr')
for d in data[2:-1:2] :
    name.append(d.find_element_by_xpath('./td[1]').text.strip('\n ')) # append name 
    dtype.append(d.find_element_by_xpath('./td[2]').text.strip('\n ')) #append dtype
    task.append(d.find_element_by_xpath('./td[3]').text.strip('\n ')) # append task
    atype.append(d.find_element_by_xpath('./td[4]').text.strip('\n ')) # append attribute type
    inst.append(d.find_element_by_xpath('./td[5]').text.strip('\n ')) # append total instances
    attri.append(d.find_element_by_xpath('./td[6]').text.strip('\n ')) # append total attributes
    year.append(d.find_element_by_xpath('./td[7]').text.strip('\n ')) # append year
    
# dataframe
df = pd.DataFrame({'Dataset Name':name,'Dataset Type':dtype,'Task':task,'Attribute Type':atype,'No. of Instances':inst,
                   'No. of Attributes':attri,'Year':year})
df
```
